# Linear Regression

- 1 trong số thuật toán cơ bản và đơn giản nhất của supervised learning
- Simple linear regression : 1 input feature
- Multiple linear regression : 2 input features trở lên
- Phù hợp với các linear dataset



Ví dụ : dự đoán mức lương của nhân viên theo  số năm kinh nghiệm

- Linear Regression đánh giá  mối quan hệ tuyến tính giữa các feature và target , đưa ra dự đoán gần với giá trị thực tế nhất có thể 



- y là giá trị muốn dự đoán (target , response , dependent variable ...)
- x là feature (independent variable , explanatory variable....)
- B0,B1,....  là hệ số (coefficient) , thể hiện mức độ đóng góp , tầm quan trọng của biến. 
- E là nhiễu 

### Ưu điểm và nhược điểm 
- Ưu điểm 
    - Đơn giản 
    - Trong những trường hợp cụ thể thì là model tốt (trường hợp dataset tuyến tính , hệ số tương quan lớn)
- Nhược điểm
    - Đơn giản quá :))
    - performance tốt trong trường hợp cụ thể và các trường hợp này ít xảy ra 
    - nhạy cảm với outliers

# Logistic Regression 

- Tên là regression nhưng lại là bài toán classification :) 
- Được xây dựng dựa trên linear regression
- Trong ***linear regression*** 
    - Y = β0 + β1x1 + β2x2 + ... + βixi + ε 
    - Y ∈ (-∞ , +∞)

- Mục tiêu Logistic Regression là ép Y ∈ [0,1] (áp dụng linear regression vào classification)
- Cách làm : 
    - cho y đi qua hàm gọi là hàm sigmoid
    - hàm sigmoid : σ(y) = ​1 / (1 + e^−y)

    

    - Khi y tiến đến -∞ thì -y tiến đến +∞ ,  e^−y tiến đến +∞ (+1 để tránh trường hợp mẫu số = 0)  ->  σ(y) = 0  

    - Không cần biết input to hay bé thì đi qua hàm sigmoid thì y sẽ chỉ nằm trong khoảng từ [0,1] , khi này chúng ta cũng có thể sử dụng để coi như là xác suất , bao nhiêu % thuộc class 0 , bao nhiêu % thuộc class 1 

- Để dễ phân biệt class 0 , 1 , chúng ta tạo ra 1 ngưỡng gọi là threshold value = 0.5 , > 0.5 thì thuộc class 1 , < 0.5 thì thuộc class 0

    

- Chúng ta có thể thay đổi threshold theo bài toán : 
    - ví dụ bài toán dự đoán covid , chúng ta có thể giảm threshold để tăng số lượng người dự đoán là có bị nhiễm covid đồng nghĩa với việc tăng tỷ lệ có thể tóm được người bị nhiễm (tăng recall) , tất nhiên chúng ta chấp nhận giảm độ chính xác (giảm precision) vì mục tiêu cần tóm được càng nhiều càng tốt (bắt nhầm hơn bỏ sót)
