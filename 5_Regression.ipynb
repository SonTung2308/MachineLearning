# Logistic Regression 

- Tn l regression nhng li l bi ton classification :) 
- c xy dng da trn linear regression
- Trong ***linear regression*** 
    - Y = 0 + 1x1 + 2x2 + ... + ixi +  
    - Y  (- , +)

- Mc tiu Logistic Regression l p Y  [0,1] (p dng linear regression vo classification)
- Cch lm : 
    - cho y i qua hm gi l hm sigmoid
    - hm sigmoid : (y) = 1 / (1 + e^y)

    

    - Khi y tin n - th -y tin n + ,  e^y tin n + (+1  trnh trng hp mu s = 0)  ->  (y) = 0  

    - Khng cn bit input to hay b th i qua hm sigmoid th y s ch nm trong khong t [0,1] , khi ny chng ta cng c th s dng  coi nh l xc sut , bao nhiu % thuc class 0 , bao nhiu % thuc class 1 

-  d phn bit class 0 , 1 , chng ta to ra 1 ngng gi l threshold value = 0.5 , > 0.5 th thuc class 1 , < 0.5 th thuc class 0

    

- Chng ta c th thay i threshold theo bi ton : 
    - v d bi ton d on covid , chng ta c th gim threshold  tng s lng ngi d on l c b nhim covid ng ngha vi vic tng t l c th tm c ngi b nhim (tng recall) , tt nhin chng ta chp nhn gim  chnh xc (gim precision) v mc tiu cn tm c cng nhiu cng tt (bt nhm hn b st)
