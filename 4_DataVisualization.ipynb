## Biểu đồ đơn biến 

### Histogram 

- Phân nhóm dữ liệu 

```python
from pandas import read_csv
from pandas.plotting import scatter_matrix 
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data = read_csv("diabetes.csv")
data.hist()
plt
```

- Nhìn vào đồ thị (histogram) ta thấy : 
    - bloodpressure và BMI có phân phối gần giống , tương tự phân phối chuẩn (gaussian)
    - isulin , diabetesPedigreeFunction , Age nhìn hao hao phân phối mũ (Exponential Distribution)
    

## Density Plot
- Tương tự histogram nhưng ở dạng đường 

```python
data.plot(kind='density',subplots=True,layout=(3,3),sharex=False)
#kind là loại 
# subplot là các plot con 
# layout là số hàng số cột cho các plot con
# sharex là chia trục (dùng chung trục)
plt
```

## Box
- Nhìn vào plot dạng box sẽ biết được dữ liệu lệch về bên nào

```python
data.plot(kind='box',subplots=True,layout=(3,3),sharex=False)
plt
```

## Correlation Matrix Plot
Cho ma trận đối xứng từ data 

```python
correlations = data.corr()
correlations
```

```python
sn.heatmap(data.corr() , annot=True)
```

## Scatter matrix Plot

- cho biết phân phối của 1 feature 

```python
scatter_matrix(data)
plt
```
