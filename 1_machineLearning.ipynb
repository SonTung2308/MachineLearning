**Cc cch tnh loss (Loss function)**

- L1 loss : Tng xch ma tr tuyt i gia sai s d on v gi tr thc t ( sau khi cng tng xch ma s chia cho s sample (nu ko chia th cng nhiu sample loss s cng ln))


- L2 loss : Tng xch ma bnh phng gia sai s d on v gi tr thc t



- Khc nhau gia L1 v L2 :
    + L2 nhy vi **sai s ln** hn L1 (sai s ^ 2 s c tp trung hn)
