**Các phương thưc học máy**
- Học Có Giám Sát (Supervised Learning)

    - Định nghĩa: Trong học có giám sát, mô hình được huấn luyện trên dữ liệu đã được gắn nhãn. Mục tiêu là học cách dự đoán nhãn từ các đặc điểm của dữ liệu.
    Ứng dụng:
        - Phân loại (Classification): Dự đoán nhãn danh mục, ví dụ như phân loại email là spam hay không spam.
        Hồi quy (Regression): Dự đoán một giá trị liên tục, ví dụ như dự đoán giá nhà.
    - Ví dụ: Mô hình dự đoán giá nhà dựa trên diện tích, vị trí, và các yếu tố khác.

- Học Không Giám Sát (Unsupervised Learning)

    - Định nghĩa: Học không giám sát không sử dụng dữ liệu đã được gắn nhãn. Thay vào đó, mô hình cố gắng tìm hiểu cấu trúc hoặc mẫu từ dữ liệu.
    - Ứng dụng:
        - Phân cụm (Clustering): Nhóm dữ liệu thành các cụm dựa trên sự giống nhau, ví dụ như phân cụm khách hàng dựa trên hành vi mua sắm.
        - Giảm chiều (Dimensionality Reduction): Giảm số lượng đặc điểm của dữ liệu mà vẫn giữ được thông tin quan trọng, ví dụ như PCA (Principal Component Analysis).
    - Ví dụ: Phân cụm khách hàng trong cơ sở dữ liệu để phát triển các chiến lược marketing hướng đích.

- Học Tăng Cường (Reinforcement Learning)

    - Định nghĩa: Trong học tăng cường, một "agent" học cách thực hiện các hành động trong một môi trường để tối đa hóa một phần thưởng cụ thể.
    - Ứng dụng:
        - Trò chơi: Đánh bại con người trong các trò chơi như cờ vua, Go.
        - Tự động hóa: Xe tự lái, robot tự động.
    - Ví dụ: Một robot học cách di chuyển trong một môi trường mô phỏng để tối ưu hóa số điểm thu được.








**Thuật ngữ:**
- Labeled data (dữ liệu có nhãn):



- 1 hàng: sample, item, resource



- 1 cột: feature, attribute




- Cột cần dự đoán: target, label



- Feature vector: tất cả giá trị liên quan tới feature



- Feature matrix (X): gồm nhiều hàng

- Label/target vector: cột giá trị cần dự đoán




**Model Training (Huấn luyện mô hình)**

- Tổng quát: Đưa vào mô hình feature vector -> giá trị dự đoán
    + Classification: đưa ra class
    + Regression: đưa ra giá trị
    + Và các tác vụ khác...

- Prediction (Giá trị dự đoán): so sánh với label (giá trị thực tế) = loss (mất mát)
- Loss càng bé càng tốt




Thực tế sẽ fit nhiều sample vào cùng 1 lúc (feature vector)




**Batch**: Số lượng phần tử được đưa vào mô hình
- thường là số mũ của 2 (4,8,16,...vv) 
- vì tổ chức thanh ghi thường là lũy thừa của 2 -> lấy batch là lũy thừa của 2 sẽ tối đa thanh ghi (ghi 64 mà lấy batch 63 thì lãng phí 1 roài)
**Loss**: So sánh giữa dự đoán và nhãn -> dùng để cập nhật lại mô hình để có dự đoán tốt hơn lần sau

**Giải thích đơn giản (giống việc học)**:
- Mô hình là mỗi chúng ta
- Feature matrix là các bài tập mà giáo viên giao
- Sau khi nhận bài tập, chúng ta sẽ ngồi giải bằng kiến thức của mình (prediction)
- Sau đó, giáo viên sẽ cung cấp cách giải và đáp án cho từng bài (label)
- Làm đúng thì điểm cao, làm sai thì điểm thấp (mô hình dự đoán sai thì loss cao, mô hình dự đoán đúng nhiều thì loss thấp)
- Sử dụng bài cô trả về sau khi chấm để tự hiểu , tự sửa sai(quá trình update)


**Loss function (hàm loss)**





**Bài toán regression**

- Giả sử chúng ta muốn xây dựng một mô hình dự đoán giá nhà ở Việt Nam:
    + 1 bộ dataset gồm nhiều hàng và 5 cột
    + 5 cột gồm: Diện tích, số tầng, thành phố, trong ngõ, giá trị căn nhà (label)

- Chia dataset: 4 cột là feature (Diện tích, số tầng, thành phố, trong ngõ), 1 cột là label (giá trị căn nhà)

- Sau khi mô hình dự đoán đưa ra prediction, chúng ta dựa vào prediction và label để tính loss




**Các cách tính loss (Loss function)**

- L1 loss : Tổng xích ma trị tuyệt đối giữa sai số dự đoán và giá trị thực tế ( sau khi cộng tổng xích ma sẽ chia cho số sample (nếu ko chia thì càng nhiều sample loss sẽ càng lớn))


- L2 loss : Tổng xích ma bình phương giữa sai số dự đoán và giá trị thực tế



- Khác nhau giữa L1 và L2 :
    + L2 nhạy với **sai số lớn** hơn L1 (sai số ^ 2 sẽ được tập trung hơn)

    


**Phân chia dataset**

- **Training set**: được sử dụng để huấn luyện mô hình.

- **Validation set**: được sử dụng trong quá trình huấn luyện tại các thời điểm nhất định (thường cuối mỗi epoch) để đảm bảo rằng mô hình đang học tốt và tránh overfitting.

- **Test set**: được sử dụng để kiểm tra khả năng tổng quát hóa của mô hình. Thường được sử dụng cuối cùng, sau khi mô hình đã được huấn luyện, để xem xét xem mô hình có hoạt động tốt với dữ liệu mới không.
