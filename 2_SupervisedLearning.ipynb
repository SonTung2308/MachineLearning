### Classification metrics

- chn dog l class c quan tm , u tin : class + 
- cc classes khc s l : class -



TP : true positives , s lng class+ (dog) prediction trng vi label hay class-




FP : false positives : s lng class+ (dog) prediction khc vi label hay class-



FN : false negatives : s lng class- (cat,..vv) prediction khc vi label  y l class+



TN: true negatives : s lng class- trng vi label



- **accuracy**
- **precision**
- **recall**
- **f1 score**



- **accuracy** : khi d liu balance (cn bng gia class+ v class -) th mi c  ngha :))
- **precison** : tp trung vo  chnh xc class+ ( cht lng) (tm c n d on th bao nhiu ci ng trong n ci tm c  )
- **recall** :  bao ph ca class+ (m rng v s lng , bt nhm hn b st :) ) (thc t c n sample thuc class+ th cta tm c bao nhiu trong n  )
- **f1 score**

**Tt c metrics ch c th nh gi performance ca model ti 1 ngng thi (threshold)**

** c 1 ci nhn tng qut v perform ca model ti nhiu ngng th chng ta s s dng 2 metrics di y**

- ROC
- AUC

**Trc ht chng ta tm hiu v 2 metrics dnh cho phng php ny**

- True Positive Rate (n chnh l recall nhng theo cch gi khc) 

- False Positive Rate

(2 mectrics ny s cng tng , cng gim)



** nh gi model , ta s a metrics ny vo h trc ta  v xt thi**




- Gii thch qua 1 cht , Ox  y l **True Positive rate** , Oy l **False Positive rate**

- Sau  ta s tng gim ngng t 0->1 

- Vi mi ngi th ngi ta s xt xem **True Positive** rate v **False Positive** rate l bao nhiu 

- Khi xt trn nhiu ngng , th ni vo s ra ng mu xanh ROC 

- ng ROC l ng cho ta thy c performace ca model ti cc ngng nh th no

- 1 m hnh tt th ng Roc phi gn im **Perfect classifier**

- **Perfect classifier** l im m **True Positive** rate l cc i = 1

- 1 m hnh t  n nm di ng mu vng , **Random classifier**

- **Random classifier** l ti ngng  , model d on trt lt , ngu nhin , 50/50 , 1 ci ng 1 ci sai

- Vy chng ta lm sao c th tm ngng na l ti u nn chn ?

- T **Perfect classifier** v 1 ng trn , im no dnh vo **ROC** th lm




**Specificity** l phn b ca **False Positive Rate** (n lin quan ti class negative nn l t ng   , mnh cng vy)

- Mc d ROC c th cho chng ta thy c Performance ca model ti cc ngng khc nhau , tuy nhin , vi nhiu model khc nhau v nhiu ng ROC khc nhau , chng ta khng th nh gi c l model no tt hn model no da trn ROC.

- C th 2 ng ROC u c 2 im gn im **Perfect classifier** ging nhau th ta s khng th bit ng ROC no tt hn

- V chng ta mun nhn vo tng th cc model ch khng ch so snh im tt nht ca tng model 

**Cch lm :** tnh din tch ca ng ROC v trc honh (Oy) , v din tch cng ln th ng ROC cng gn im **Perfect classifier** (chng ta c th chn c ng ROC no tt hn ri) -> chnh l AUC **Area under the ROC curve**

- uhmmm , tuy nhin th , trng hp AUC ca model 1 tt hn AUC ca model 2 nhng model 2 c ROC cho 1 im gn **Perfect classifier** l bnh thng , lc ny th chng ta cn xem xt l model cn quan tm ti 1 ngng c th hn hay model tng qut hn 

- v d model t bin ng (khng phi thay i ngng nhiu) th quan ti model no cho ROC c im gn **Perfect classifier** hn v ngc li

- n gin hn classification : thng ch dng 2 metrics l

- RMSE : cn bc hai ca trung bnh ca bnh phng cc sai s



- S khc bit chnh: RMSE phn nh cm nhn v sai s ln tt hn do n bnh phng sai s trc khi tnh trung bnh, trong khi MAE cung cp mt cch nh gi trc tip v khng "pht" sai s ln.
