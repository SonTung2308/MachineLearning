### Classification: dog vs cat



### Classification metrics

- chọn dog là class được quan tâm , ưu tiên : class + 
- các classes khác sẽ là : class -



TP : true positives , số lượng class+ (dog) prediction trùng với label hay class-




FP : false positives : số lượng class+ (dog) prediction khác với label hay class-



FN : false negatives : số lượng class- (cat,..vv) prediction khác với label ở đây là class+



TN: true negatives : số lượng class- trùng với label



- **accuracy**
- **precision**
- **recall**
- **f1 score**



- **accuracy** : khi dữ liệu balance (cân bằng giữa class+ và class -) thì mới có ý nghĩa :))
- **precison** : tập trung vào độ chính xác class+ ( chất lượng) (tìm được n dự đoán thì bao nhiêu cái đúng trong n cái tìm được đó )
- **recall** : độ bao phủ của class+ (mở rộng về số lượng , bắt nhầm hơn bỏ sót :) ) (thực tế có n sample thuộc class+ thì cta tìm đựơc bao nhiêu trong n đó )
- **f1 score**



**Ví dụ đi :** 

- chúng ta cần 1 model để dự đoán người mắc covid đi , câu hỏi đặt ra là , chúng ta tập trung vào precision hay recall ? 

- tất nhiên là recall đúng không , vì chúng ta cần bắt cả đám f1, f2 , f3 ... chứ không chỉ riêng f0. tóm cổ cả những đứa có nguy cơ f0 đi cách ly chứ 



**Ví dụ tiếp**
- chúng ta đi hẹn hò với crush đi , mời crush đi ăn, cta cần tìm 1 list danh sách các quán ăn, vậy trong list đó chúng ta sẽ chú tâm vào quán ăn ngon hay số lượng quán ăn ngon nào ?

**Chú ý**

**Tất cả metrics chỉ có thể đánh giá performance của model tại 1 ngưỡng thôi (threshold)**

**Để có 1 cái nhìn tổng quát về perform của model tại nhiều ngưỡng thì chúng ta sẽ sử dụng 2 metrics dưới đây**

- ROC
- AUC

**Trước hết chúng ta tìm hiểu về 2 metrics dành cho phương pháp này**

- True Positive Rate (nó chính là recall nhưng theo cách gọi khác) 

- False Positive Rate

(2 mectrics này sẽ cùng tăng , cùng giảm)



**Để đánh giá model , ta sẽ đưa metrics này vào hệ trục tọa độ và xét thôi**




- Giải thích qua 1 chút , Ox ở đây là **True Positive rate** , Oy là **False Positive rate**

- Sau đó ta sẽ tăng giảm ngưỡng từ 0->1 

- Với mỗi ngưỡi thì người ta sẽ xét xem **True Positive** rate và **False Positive** rate là bao nhiêu 

- Khi xét trên nhiều ngưỡng , thì nối vào sẽ ra đường màu xanh ROC 

- Đường ROC là đường cho ta thấy được performace của model tại các ngưỡng như thế nào

- 1 mô hình tốt thì đường Roc phải gần điểm **Perfect classifier**

- **Perfect classifier** là điểm mà **True Positive** rate là cực đại = 1

- 1 mô hình tệ à nó nằm dưới đường màu vàng , **Random classifier**

- **Random classifier** là tại ngưỡng đó , model dự đoán trật lất , ngẫu nhiên , 50/50 , 1 cái đúng 1 cái sai

- Vậy chúng ta làm sao có thể tìm ngưỡng naò là tối ưu nên chọn ?

- Từ **Perfect classifier** vẽ 1 đường tròn , điểm nào dính vào **ROC** thì lụm




**Specificity** là phần bù của **False Positive Rate** (nó liên quan tới class negative nên là ít ng để ý , mình cũng vậy)

- Mặc dù ROC có thể cho chúng ta thấy được Performance của model tại các ngưỡng khác nhau , tuy nhiên , với nhiều model khác nhau và nhiều đường ROC khác nhau , chúng ta không thể đánh giá được là model nào tốt hơn model nào dựa trên ROC.

- Có thể 2 đường ROC đều có 2 điểm gần điểm **Perfect classifier** giống nhau thì ta sẽ không thể biết đường ROC nào tốt hơn

- Và chúng ta muốn nhìn vào tổng thể các model chứ không chỉ so sánh điểm tốt nhất của từng model 

**Cách làm :** tính diện tích của đường ROC và trục hoành (Oy) , vì diện tích càng lớn thì đường ROC càng gần điểm **Perfect classifier** (chúng ta có thể chọn được đường ROC nào tốt hơn rồi) -> chính là AUC **Area under the ROC curve**

- uhmmm , tuy nhiên thì , trường hợp AUC của model 1 tốt hơn AUC của model 2 nhưng model 2 có ROC cho 1 điểm gần **Perfect classifier** là bình thường , lúc này thì chúng ta cần xem xét là model cần quan tâm tới 1 ngưỡng cụ thể hơn hay model tổng quát hơn 

- ví dụ model ít biến động (không phải thay đổi ngưỡng nhiều) thì quan tới model nào cho ROC có điểm gần **Perfect classifier** hơn và ngược lại

## Đánh giá mô hình cho Regression




- Đơn giản hơn classification : thường chỉ dùng 2 metrics là

- MAE : trung bình của giá trị tuyệt đối của các sai số. Sai số ở đây là sự chênh lệch giữa giá trị dự đoán và giá trị thực tế.
- RMSE : căn bậc hai của trung bình của bình phương các sai số



- Sự khác biệt chính: RMSE phản ánh cảm nhận về sai số lớn tốt hơn do nó bình phương sai số trước khi tính trung bình, trong khi MAE cung cấp một cách đánh giá trực tiếp và không "phạt" sai số lớn.
- Lựa chọn:
    - Nếu các sai số lớn là không mong muốn và cần được "phạt" nặng, RMSE là sự lựa chọn tốt.
    - Nếu bạn muốn đánh giá mô hình một cách đơn giản và trực tiếp, hoặc trong các trường hợp mà sai số lớn không quá quan trọng, MAE có thể phù hợp hơn.
